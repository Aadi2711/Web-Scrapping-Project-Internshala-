```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
final= pd.DataFrame()

for j in range(1,21):
    url= 'https://internshala.com/internships/page-{}/'.format(j)
    webpage=requests.get(url).text
    soup=BeautifulSoup(webpage,'lxml')
    internships=soup.find_all('div',class_='internship_meta')
    
    opportunity_name=[]
    company_name=[]
    location=[]
    duration=[]
    stipend=[]

    for i in range(2,len(internships)):
        opportunity_name.append((internships[i].find('a',class_='view_detail_button').text.strip()))
        company_name.append(internships[i].find('a', class_='link_display_like_text view_detail_button').text.strip())
        location.append(internships[i].find('a',class_='location_link view_detail_button').text.strip())
        duration.append(internships[i].find_all('div', class_='item_body')[1].text.strip())
        stipend.append(internships[i].find_all('div', class_='item_body')[2].text.strip())
        
    data={'Internship':opportunity_name,'Company':company_name,'Location':location,'Duration':duration,'Stipend':stipend}
    df= pd.DataFrame(data)
    final=pd.concat([final,df], ignore_index=True)
```

```python
copy= final.copy()
```

```python
def numeric(string):
    ans=""
    for i in string:
        if i.isnumeric() or i=="-":
            ans+=i
    return ans
```

```python
final['Stipend']=final['Stipend'].apply(numeric)
```
